```python
#rename columns
df2 = df2.rename(columns = {
                'Year of event': 'year',
                'Event dates': 'race_day',
                 'Event name': 'race_name', 
                 'Event distance/length': 'race_length', 
                 'Event number of finishers': 'race_no_of_finishers' ,
                 'Athlete performance': 'athlete_performance', 
                 'Athlete gender': 'athlete_gender',
                 'Athlete average speed': 'athlete_average_speed', 
                 'Athlete ID': 'athlete_id', 
                 'athlete age':'athlete_age' 
})
df2
```

```python
#reorder columns

df3 = df2[['race_day','race_name','race_length', 'athlete_id', 'athlete_gender','athlete_age', 'athlete_performance','athlete_average_speed', 'race_no_of_finishers','year']]
```

```python
sns.histplot(df3['race_length'])
plt
```

```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed',hue = 'athlete_gender')

plt.title('Athlete Average Speed Vs Athlete Age & Gender', color = 'red')
plt.xlabel('Athlete Age')
plt.ylabel('Athlete Average Speed')
plt.show
```
